```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('chronic_kidney_disease.csv')
```

```python
df
```

```python
df.info()
```

```python
df = df.replace("?", np.NaN) 
df
```

```python
mask = df.isna().sum() / len(df) < .2
df_reduced = df.loc[:, mask]
df_reduced
```

```python
reduced_ckd = df_reduced.dropna()
reduced_ckd
```

```python
for col in ['Age', 'Blood Pressure', 'Blood Glucose Random', 'Packed Cell Volume']:
    reduced_ckd[col] = reduced_ckd[col].astype(np.int16)

for col in ['Specific Gravity', 'Serum Creatinine', 'Hemoglobin', 'Blood Urea']:
    reduced_ckd[col] = reduced_ckd[col].astype(np.float16)

for col in ['Albumin', 'Sugar', 'Pus Cell', 'Pus Cell clumps', 'Bacteria', 'Hypertension', 'Diabetes Mellitus', 'Coronary Artery Disease', 'Appetite', 'Pedal Edema', 'Anemia', 'Class']:
    reduced_ckd[col] = reduced_ckd[col].astype('category')
```

```python
numeric_ckd = reduced_ckd.select_dtypes(include='number')
numeric_ckd
```

```python
numeric_ckd.describe()
```

```python
from sklearn.feature_selection import VarianceThreshold
```

```python
sel = VarianceThreshold(threshold=.05)
sel.fit(numeric_ckd) #/numeric_ckd.mean() ?
mask = sel.get_support() 
numeric_ckd_reduced = numeric_ckd.loc[:, mask]
numeric_ckd_reduced
```
