```python

from itertools import combinations

# Define node_in_open_triangle()
def node_in_open_triangle(G, n):
    in_open_triangle = False
    
    # Iterate over all possible triangle relationship combinations
    for n1, n2 in combinations(G.neighbors(n), 2):
        # Check if n1 and n2 do NOT have an edge between them
        if not G.has_edge(n1, n2):
            in_open_triangle = True
            break
    return in_open_triangle
    
# Compute the number of open triangles in T
num_open_triangles = 0

# Iterate over all the nodes in T
for n in T.nodes():
    # Check if the current node is in an open triangle
    if node_in_open_triangle(T, n):
        # Increment num_open_triangles
        num_open_triangles += 1
print(num_open_triangles)
```

```python
# Define maximal_cliques()
def maximal_cliques(G, size):
    """
    Finds all maximal cliques in graph `G` that are of size `size`.
    """
    mcs = []
    # Use find_cliques_recursive(G) function to find the maximal cliques in a directed graph
    #what about find_cliques(G)?
    for clique in nx.find_cliques_recursive(G): 
        if len(clique) == size:     # Compare if the len(clique) if of size 'size', if so append(clique) to the list mcs
            mcs.append(clique)
    return mcs
    
print(len(maximal_cliques(T, 3)))
```

```python
#Bipartite Graphs
# Read in the data from clubs_people (list of tuples representing the edge list)
#.add_edges_from(clubs_people)
B = nx.Graph()
B.add_edges_from(clubs_people)

#print(B.nodes())
# Assign nodes to 'clubs' or 'people' partitions
for n, d in B.nodes(data=True):
    #print(n)
    #print(d)
    if '.' in n:
        B.nodes[n]['bipartite'] = 'people'
    else:
        B.nodes[n]['bipartite'] = 'clubs'
        
# Print the edges of the graph
print(B.edges())
```

```python

people = [n for n in B.nodes() if B.nodes[n]['bipartite'] == 'people']
clubs = [n for n in B.nodes() if B.nodes[n]['bipartite'] == 'clubs']
```

```python
print(clubs)
```

```python

peopleB = nx.bipartite.projected_graph(B, people)
clubsB = nx.bipartite.projected_graph(B, clubs)
```

```python

# Plot the degree centrality (dc) distribution of both node partitions from the original graph
plt.figure()
original_dc = nx.bipartite.degree_centrality(B, people) #clubs?
# Remember that you can directly plot dictionary values.
plt.hist(original_dc, alpha=0.5)
plt.yscale('log')
plt.title('Bipartite degree centrality')
plt.show()
```

```python
# Plot the degree centrality distribution of the peopleB graph
plt.figure()
people_dc = nx.degree_centrality(peopleB)
plt.hist(people_dc.values())
plt.yscale('log')
plt.title('Degree centrality of people partition')
plt.show()
```
